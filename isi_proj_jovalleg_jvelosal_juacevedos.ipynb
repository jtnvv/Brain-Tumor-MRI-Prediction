## **A. Descripción del problema: Detección y Clasificación de Tumores Cerebrales por MRI**

###  Introducción

Las imágenes por resonancia magnética (RM) son una herramienta fundamental en el diagnóstico médico moderno, especialmente en neurología. A diferencia de otras modalidades de imagen que utilizan radiación ionizante, la RM emplea un potente campo magnético y ondas de radio para generar imágenes detalladas de órganos y tejidos blandos del cuerpo [1]. Esta capacidad de visualización no invasiva y de alta resolución la convierte en la técnica de elección para la evaluación del cerebro, permitiendo identificar estructuras anatómicas con gran precisión y detectar anomalías como tumores. En el ámbito clínico, las RM cerebrales son esenciales para el diagnóstico y seguimiento de diversas afecciones neurológicas, proporcionando información crucial para la toma de decisiones médicas.

El análisis de estas imágenes, sin embargo, es un proceso intensivo que requiere la experiencia de radiólogos y neurólogos, lo que puede generar una carga significativa de trabajo y, en ocasiones, retrasos en el diagnóstico, especialmente en regiones con escasez de especialistas. Para abordar este desafío, nuestro proyecto se centra en el desarrollo de un sistema de clasificación automática de imágenes de RM cerebrales. Este sistema busca clasificar las imágenes en cuatro categorías distintas: Glioma, Meningioma, Pituitario y No tumor, con el objetivo de optimizar los tiempos de diagnóstico y reducir la dependencia de la intervención manual intensiva. La naturaleza de esta tarea implica un aprendizaje supervisado, donde el modelo se entrena con un conjunto de datos etiquetados para aprender a mapear características de las imágenes a las categorías de diagnóstico [2]. Es importante destacar que, dada la aplicación médica, las consecuencias de un diagnóstico erróneo ya sea un falso positivo (detectar un tumor donde no lo hay) o un falso negativo (no detectar un tumor presente) pueden ser graves. Por ello, el desarrollo de un modelo con alta sensibilidad y precisión es de vital importancia.

### Glosario
- *Aprendizaje Supervisado (Supervised Learning)*: Es un método de aprendizaje automático donde un algoritmo aprende a predecir resultados a partir de un conjunto de datos que ya están etiquetados.

- *Clasificación Automática (Automatic Classification)*: Se refiere a la capacidad de un sistema de inteligencia artificial para asignar categorías predefinidas a nuevos datos sin intervención humana.

- *Conjunto de Datos (Dataset)*: Es una colección organizada de datos, como imágenes médicas, utilizada para entrenar, validar y probar modelos de aprendizaje automático.

- *Falsos Negativos (False Negatives)*: Ocurre cuando un modelo de clasificación predice incorrectamente la ausencia de una condición que en realidad está presente, como no detectar un tumor existente.

- *Falsos Positivos (False Positives)*: Se produce cuando un modelo de clasificación predice incorrectamente la presencia de una condición que en realidad está ausente, como diagnosticar un tumor donde no lo hay.

- *GPU (Graphics Processing Unit)*: Es un procesador especializado diseñado para acelerar cálculos paralelos, crucial para el entrenamiento eficiente de modelos de aprendizaje profundo.

- *Imágenes por Resonancia Magnética (MRI - Magnetic Resonance Imaging)*: Es una técnica de diagnóstico por imagen no invasiva que usa campos magnéticos y ondas de radio para crear imágenes detalladas de órganos y tejidos internos.

- *Modelo de Clasificación (Classification Model)*: Es un algoritmo de aprendizaje automático entrenado para asignar una categoría o clase específica a una nueva entrada de datos.

- *Sobreentrenamiento (Overfitting)*: Es un fenómeno en el aprendizaje automático donde un modelo aprende demasiado los detalles y el ruido de los datos de entrenamiento, perdiendo su capacidad para generalizar a datos nuevos.

- *Transfer Learning (Aprendizaje por Transferencia)*: Es una técnica que reutiliza un modelo pre-entrenado en una tarea grande y similar como punto de partida para una nueva tarea específica, aprovechando el conocimiento ya adquirido.

### Desafios
El desarrollo de modelos de clasificación de imágenes médicas presenta desafíos inherentes que deben ser cuidadosamente abordados para garantizar un rendimiento robusto y confiable. Uno de los problemas más comunes y críticos es el sobreentrenamiento. Este fenómeno ocurre cuando un modelo aprende en exceso los detalles y el ruido del conjunto de datos de entrenamiento, hasta el punto de memorizarlos en lugar de aprender patrones generalizables [3]. Como resultado, el modelo exhibe un rendimiento excelente en los datos de entrenamiento, pero falla estrepitosamente al enfrentarse a imágenes nuevas y no vistas, lo que se traduce en una baja capacidad de generalización. Para mitigar el sobreentrenamiento, es crucial emplear técnicas como la regularización (por ejemplo, dropout), el aumento de datos (data augmentation), y el uso de conjuntos de validación independientes para monitorear el rendimiento del modelo durante el entrenamiento y detenerlo cuando comienza a sobreajustarse.

Además del sobreentrenamiento, los desafíos de hardware representan una barrera significativa, especialmente al trabajar con conjuntos de datos de imágenes grandes y modelos de aprendizaje profundo complejos. El entrenamiento de redes neuronales convolucionales (CNNs) de última generación, que son la base para la clasificación de imágenes médicas, demanda una considerable capacidad computacional, incluyendo procesadores de alto rendimiento (GPUs) y una gran cantidad de memoria RAM [4]. La ausencia de este hardware especializado puede prolongar drásticamente los tiempos de entrenamiento, limitar la complejidad de los modelos que se pueden explorar e incluso impedir la implementación de arquitecturas más avanzadas que podrían ofrecer un mejor rendimiento. Estos desafíos pueden restringir la viabilidad del proyecto, especialmente en entornos con recursos limitados, lo que impulsa la búsqueda de soluciones eficientes que minimicen la demanda computacional sin comprometer la precisión.

### Alternativas de Modelos de Clasificación
En el ámbito de la clasificación de imágenes, existen principalmente dos enfoques para el desarrollo de modelos de aprendizaje profundo: construir una arquitectura desde cero o utilizar un enfoque de transfer learning. La creación de un modelo desde cero implica diseñar y entrenar una red neuronal convolucional completamente nueva utilizando únicamente el conjunto de datos disponible para el proyecto [5]. La ventaja principal de este enfoque es la flexibilidad, ya que la arquitectura puede ser adaptada específicamente a las características del problema y los datos. Sin embargo, este método requiere un volumen muy grande de datos etiquetados para entrenar la red de manera efectiva y evitar el sobreajuste. Además, el entrenamiento de una red desde cero es computacionalmente intensivo y puede llevar un tiempo considerable, lo que lo hace menos práctico para proyectos con recursos limitados o conjuntos de datos de tamaño moderado.

Por otro lado, el transfer learning aprovecha el conocimiento adquirido por modelos pre-entrenados en tareas similares, generalmente en grandes conjuntos de datos de imágenes como ImageNet. En nuestro proyecto, adoptamos este enfoque utilizando arquitecturas como ResNet50 y EfficientNetB4. La ventaja fundamental del transfer learning es que los modelos pre-entrenados ya han aprendido una vasta jerarquía de características visuales, desde bordes y texturas básicas hasta formas más complejas. Esto significa que no es necesario entrenar el modelo desde cero, sino que se pueden adaptar las capas finales a la tarea específica de clasificación de tumores cerebrales. ResNet50 es conocida por su capacidad para manejar el problema del gradiente evanescente mediante conexiones residuales, permitiendo la construcción de redes muy profundas [6]. EfficientNetB4 destaca por su eficiencia y su enfoque compuesto de escalado, que optimiza simultáneamente la profundidad, el ancho y la resolución de la red para un mejor rendimiento.

La elección del transfer learning es superior en nuestro contexto por varias razones. Primero, reduce drásticamente la cantidad de datos de entrenamiento requeridos, lo cual es ventajoso ya que los conjuntos de datos de imágenes médicas pueden ser limitados. Segundo, acelera significativamente el proceso de entrenamiento, ya que gran parte de los pesos del modelo ya están inicializados de manera efectiva. Finalmente, los modelos pre-entrenados han demostrado su robustez y capacidad para extraer características relevantes, lo que generalmente conduce a un mejor rendimiento y mayor precisión en comparación con los modelos entrenados desde cero en conjuntos de datos más pequeños. Este enfoque nos permite construir un sistema sensible y preciso de manera más eficiente, lo cual es crucial dada la criticidad del diagnóstico de tumores cerebrales.

### Preprocesamiento de imagenes: ResNet50
- *Conversión a RGB*

  Asegura que todas las imágenes tengan tres canales, lo cual es necesario porque ResNet50 espera entradas en formato RGB. Esto es importante especialmente en datasets médicos donde pueden existir imágenes en escala de grises.

- *Redimensionamiento*

  ResNet50 requiere imágenes de un tamaño específico como entrada. Este tamaño es típicamente 224x224 píxeles, por lo que redimensionar garantiza compatibilidad con la arquitectura y permite el procesamiento en batch.

- *Conversión a array (NumPy)*

  Transforma la imagen PIL a una matriz NumPy, que es el formato compatible con TensorFlow/Keras. Esta representación es fundamental para el procesamiento numérico eficiente.

- *Preprocesamiento específico del modelo*

  Esto normaliza los valores de píxeles según los estándares del modelo preentrenado (en este caso ResNet50). Normalizar las imágenes mejora la convergencia del modelo durante el entrenamiento y evita problemas derivados de rangos de intensidad inconsistentes.

- *Validación básica (detección de outliers)*  

  `if np.max(image) > 1e6 or np.isnan(image).any()`  
  Esto filtra imágenes que puedan haber sido mal leídas o que generen errores numéricos durante el entrenamiento. Esto ayuda a mantener un conjunto limpio, libre de anomalías que podrían sesgar el proceso de aprendizaje.

- *Asignación de etiquetas (mapeo de clases)*  

  `label_to_idx = {label: idx ...}`  
  Convierte etiquetas de texto a valores numéricos enteros, necesarios para el entrenamiento supervisado en frameworks como TensorFlow.

- *Almacenamiento en disco (como .npy)*

  Guardar los arrays preprocesados evita reprocesar todas las imágenes cada vez que se ejecuta el modelo, lo cual ahorra tiempo y recursos computacionales, especialmente en entornos como Google Colab.

#### Test Data ResNet50

````python
import os
import numpy as np
from PIL import Image, UnidentifiedImageError
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input

# ------------------------
# CONFIGURACIÓN
# ------------------------
TEST_IMAGES_PATH = "/content/drive/Shareddrives/ISI/test"
IMAGE_SIZE = (224, 224)
SAVE_PATH = "/content/drive/Shareddrives/ISI/preprocessed_data"

os.makedirs(SAVE_PATH, exist_ok=True)

# ------------------------
# FUNCIÓN DE PREPROCESAMIENTO GENERAL
# ------------------------
def preprocess_test_images(image_path, image_size, preprocess_function=None, enhance_function=None):
    """
    Procesa imágenes de test, redimensiona, convierte a array y aplica preprocesamiento
    específico del modelo si se proporciona.
    """
    X_test = []
    ids_test = []
    bad_images = []

    print(f"Procesando imágenes de test desde {image_path}...")

    image_files = sorted(os.listdir(image_path))

    for i, img_filename in enumerate(image_files):
        if (i + 1) % 100 == 0:
            print(f"Procesando imagen {i + 1}/{len(image_files)}: {img_filename}")

        if not img_filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        img_full_path = os.path.join(image_path, img_filename)

        try:
            image = Image.open(img_full_path)

            if image.mode != 'RGB':
                image = image.convert('RGB')

            image = image.resize(image_size)
            image_array = img_to_array(image)

            if enhance_function:
                image_array = enhance_function(image_array)

            if preprocess_function:
                image_array = preprocess_function(image_array)

            if np.max(image_array) > 1e6 or np.isnan(image_array).any():
                print(f"Skipping image {img_filename} due to potential outlier values after preprocessing.")
                bad_images.append(img_filename)
                continue

            X_test.append(image_array)
            ids_test.append(img_filename)

        except (UnidentifiedImageError, FileNotFoundError, OSError) as e:
            print(f"Skipping image {img_filename} due to error: {e}")
            bad_images.append(img_filename)

    print(f"Finished processing images.")
    print(f"Total valid test images processed: {len(X_test)}")
    print(f"Images skipped due to errors or outliers: {len(bad_images)}")

    return np.array(X_test), np.array(ids_test)


# ------------------------
# PROCESAR IMÁGENES DE TEST PARA ResNet50
# ------------------------
# Use the general function for ResNet50 test data
X_test, ids_test = preprocess_test_images(
    TEST_IMAGES_PATH,
    IMAGE_SIZE,
    preprocess_function=preprocess_input
)


# ------------------------
# GUARDAR ARRAYS
# ------------------------
x_test_save_path = os.path.join(SAVE_PATH, "X_test.npy")
ids_test_save_path = os.path.join(SAVE_PATH, "ids_test.npy")

print(f"Saving test arrays to {SAVE_PATH}...")
np.save(x_test_save_path, X_test)
np.save(ids_test_save_path, ids_test)
print("Test arrays saved successfully.")
````

### Preprocesamiento de imagenes: EfficientNet_B4
Estas fueron los puntos principales para la preparación de los datos para entrenar el modelo, y para test

1. Normalización de intensidades
Aunque el preprocesamiento no normaliza directamente los valores de los píxeles en el script principal, esta técnica se aplica de forma explícita durante el entrenamiento del modelo. Específicamente, se utiliza la transformación transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]), que corresponde a la media y desviación estándar del conjunto de datos ImageNet, sobre el cual EfficientNet-B4 fue preentrenado. Esta normalización se aplica después de convertir las imágenes a tensores (ToTensor()), lo que garantiza que los valores estén en el rango [0, 1] antes de estandarizar. Esta decisión es adecuada, ya que permite mantener las imágenes en su formato original (uint8) en disco y realizar la normalización al vuelo, aprovechando la eficiencia y flexibilidad del DataLoader de PyTorch.

2. Redimensionamiento uniforme
El redimensionamiento de las imágenes se realiza tanto durante el preprocesamiento como durante el entrenamiento. En el script de preprocesamiento, las imágenes se redimensionan a un tamaño fijo de 380×380 píxeles utilizando el método Image.LANCZOS, que es una técnica de interpolación de alta calidad capaz de conservar detalles relevantes. Esta operación asegura que todas las imágenes tengan un tamaño uniforme, condición necesaria para ser procesadas por la arquitectura EfficientNet-B4. Además, el mismo tamaño se refuerza durante la etapa de entrenamiento mediante transforms.Resize((380, 380)), garantizando consistencia en las dimensiones de entrada del modelo. Esta uniformidad es esencial para evitar errores estructurales durante la inferencia y entrenamiento, y para asegurar una representación espacial coherente.

3. Conversión de escala de grises o RGB estándar
Dado que algunas imágenes en el conjunto pueden estar en escala de grises u otros formatos no compatibles con redes convolucionales preentrenadas, se realiza una conversión explícita al formato RGB durante el preprocesamiento. Esto se implementa mediante img.convert("RGB") cuando el modo de la imagen no es RGB. Además, luego de aplicar CLAHE en escala de grises, se vuelve a convertir el resultado a RGB utilizando OpenCV: cv2.cvtColor(enhanced, cv2.COLOR_GRAY2RGB). Esta reconversión es fundamental, ya que modelos como EfficientNet-B4 esperan tres canales de entrada. La conversión garantiza que incluso imágenes originalmente monocromáticas puedan ser interpretadas correctamente por la red sin modificar la estructura del modelo.

4. Eliminación de ruido
En este caso, no se aplican filtros tradicionales de reducción de ruido como el filtro gaussiano o mediana. Sin embargo, se utiliza una técnica más avanzada y especializada: CLAHE (Contrast Limited Adaptive Histogram Equalization). CLAHE mejora el contraste local de las imágenes médicas sin amplificar excesivamente el ruido, gracias al parámetro clipLimit=1.5, que controla la limitación del histograma. Esta técnica es particularmente útil en imágenes médicas donde estructuras de interés pueden tener bajo contraste con respecto al fondo. Aunque su propósito principal no es la reducción de ruido, CLAHE cumple una función similar en la práctica, al suavizar regiones homogéneas y evitar la aparición de artefactos. Por lo tanto, se considera que la eliminación de ruido se encuentra parcialmente cubierta por el uso estratégico de CLAHE.

- Conversión a RGB

  En entornos médicos, muchas imágenes se capturan en escala de grises, especialmente en modalidades como resonancia magnética o rayos X. Convertir estas imágenes a formato RGB permite que puedan ser utilizadas por modelos preentrenados como ResNet50, que fueron diseñados para trabajar con imágenes naturales a color. Esta conversión no agrega información clínica adicional, pero permite la compatibilidad técnica con la arquitectura sin afectar los patrones estructurales que el modelo debe aprender.

- Redimensionamiento

  Las imágenes médicas pueden variar en tamaño dependiendo del equipo y del protocolo de adquisición. Redimensionarlas a un formato estándar como 224x224 píxeles asegura que la información clínica relevante (como masas, lesiones o áreas afectadas) sea procesada de forma uniforme. Aunque puede implicar una ligera pérdida de resolución, este paso es necesario para que el modelo opere eficientemente sin comprometer de forma significativa la detección de patrones diagnósticos generales.

- Validación básica (detección de outliers)

  En datasets clínicos, pueden existir imágenes mal procesadas, corrompidas o vacías que no aportan información útil y pueden inducir errores graves en el modelo. Detectar estas imágenes y excluirlas garantiza que el entrenamiento se base únicamente en datos válidos, lo que mejora la seguridad y la calidad del sistema de ayuda al diagnóstico.

- Asignación de etiquetas (mapeo de clases)

  Asociar imágenes a etiquetas numéricas facilita la clasificación automática. Por ejemplo, convertir etiquetas como "glioma", "meningioma" o "no tumoral" a valores enteros permite que el modelo aprenda a diferenciar condiciones médicas con precisión. Este paso es fundamental para entrenar modelos de diagnóstico supervisado.

  Optimiza el espacio, lo que lo hace mucho mas practico en entornos medicos

### ***ResNet50***



![image.png](https://i.ytimg.com/vi/woEs7UCaITo/maxresdefault.jpg)

ResNet (Red Residual) es una arquitectura de redes neuronales convolucionales profundas diseñada para facilitar el entrenamiento de modelos con muchas capas. Su principal innovación es el uso de conexiones residuales, que permiten que la información fluya directamente entre capas no contiguas. Esto se representa en la imagen por medio de las líneas curvas con sumas, donde la salida de una capa se suma directamente a la salida de otra más adelante, evitando así el paso por capas intermedias. Esta estrategia ayuda a mitigar problemas comunes en redes profundas, como la desaparición del gradiente y la degradación del desempeño a medida que se agregan más capas.

En el flujo mostrado, la imagen de entrada pasa por una serie de convoluciones de tamaño 3x3 con distintos números de filtros (64, 128, 256, 512), donde se extraen progresivamente características más complejas. Luego, la salida se aplanada y se pasa por una capa completamente conectada de 1000 dimensiones que produce la predicción final. Gracias a su diseño, ResNet permite entrenar redes profundas con mayor estabilidad y precisión, siendo una arquitectura ampliamente utilizada en tareas de clasificación de imágenes, incluyendo aplicaciones médicas y científicas.

### ***EfficientNet-B4***

Como se muestra en la siguiente figura, la imagen de entrada (por ejemplo, un carro) pasa a través de una serie de bloques representados por rectángulos azules y verdes. Estos bloques corresponden a capas convolucionales que extraen y procesan características visuales en distintos niveles. EfficientNet se caracteriza por escalar de manera balanceada tres dimensiones clave del modelo: profundidad (más capas), anchura (más canales por capa) y resolución (imágenes de entrada más grandes). A diferencia de otras arquitecturas que escalan estas dimensiones de forma arbitraria, EfficientNet lo hace de manera conjunta y eficiente, logrando una mayor precisión con menos parámetros y menor carga computacional.

#### Configuración

````python
#Ruta utilizada en la VM
PREPROCESSED_PATH = Path('./')
BATCH_SIZE = 8
LR = 8e-5
LABEL_SMOOTH = 0.05
DROPOUT = 0.6
RUNS = 50
NUM_EPOCHS = 50
PATIENCE_ES = 7
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Usando dispositivo: {device}, tasa de aprendizaje inicial: {LR:.1e}")

os.makedirs(PREPROCESSED_PATH / 'models', exist_ok=True)
os.makedirs(PREPROCESSED_PATH / 'results', exist_ok=True)
````

#### Funciones metricas

````python
def compute_metrics(y_true, y_pred, y_probs, class_names, save_prefix=''):
    acc = accuracy_score(y_true, y_pred)
    recall_per_class = recall_score(y_true, y_pred, average=None, zero_division=0)
    recall_macro = recall_score(y_true, y_pred, average='macro', zero_division=0)
    composite_score = (recall_macro + acc) / 2

    print("\n--- MÉTRICAS ---")
    print(f"Accuracy:        {acc:.4f}")
    print(f"Recall macro:    {recall_macro:.4f}")
    print(f"Composite score: {composite_score:.4f}")
    for i, r in enumerate(recall_per_class):
        print(f"Recall clase {class_names[i]}: {r:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Matriz de Confusión")
    plt.tight_layout()
    plt.savefig(f"{save_prefix}confusion_matrix.png")
    plt.close()

    y_true_bin = F.one_hot(torch.tensor(y_true), num_classes=len(class_names)).numpy()
    y_probs_np = np.array(y_probs)
    plt.figure(figsize=(10, 8))
    for i in range(len(class_names)):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_probs_np[:, i])
        auc = roc_auc_score(y_true_bin[:, i], y_probs_np[:, i])
        plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC={auc:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("Curvas ROC por Clase")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.savefig(f"{save_prefix}roc_curves.png")
    plt.close()
````

#### Dataset

````python
class NumpyDataset(Dataset):
    def __init__(self, X_path, y_path=None, label_map_path=None, transform=None):
        self.X = np.load(X_path)
        self.transform = transform
        self.y = np.load(y_path) if y_path is not None else None
        if label_map_path:
            mapping = np.load(label_map_path, allow_pickle=True).item()
            self.idx_to_label = mapping['idx_to_label']
        else:
            self.idx_to_label = None

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        img = self.X[idx]
        if self.transform:
            img = self.transform(img)
        return (img, int(self.y[idx])) if self.y is not None else img
````

#### Transformaciones

````python
train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomResizedCrop(380, scale=(0.85, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.GaussianBlur(kernel_size=3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
val_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((380, 380)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

tta_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(410),
    transforms.FiveCrop(380),
    transforms.Lambda(lambda crops: torch.stack([
        transforms.ToTensor()(crop) for crop in crops])),
    transforms.Lambda(lambda crops: transforms.Normalize(
        [0.485, 0.456, 0.406],
        [0.229, 0.224, 0.225])(crops))
])
````

#### Carga de datos

````python
X_path = PREPROCESSED_PATH / 'X_B4_enhanced.npy'
y_path = PREPROCESSED_PATH / 'y_B4_enhanced.npy'
label_map_path = PREPROCESSED_PATH / 'label_mappings_B4.npy'
mapping = np.load(label_map_path, allow_pickle=True).item()['idx_to_label']
dataset = NumpyDataset(X_path, y_path, label_map_path, transform=train_transform)
NUM_CLASSES = len(mapping)
total = len(dataset)
val_size = int(0.15 * total)
train_size = total - val_size
train_indices, val_indices = random_split(range(total), [train_size, val_size])
ids = np.load(PREPROCESSED_PATH / 'ids_test_B4.npy')
````

#### Bucle principal

````python
for run in range(1, RUNS + 1):
    print(f"\n=== Run {run}/{RUNS} (B4) ===")

    train_ds = torch.utils.data.Subset(dataset, train_indices)
    val_ds   = torch.utils.data.Subset(dataset, val_indices)
    train_ds.dataset.transform = train_transform
    val_ds.dataset.transform   = val_transform
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
    val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

    model = EfficientNet.from_pretrained('efficientnet-b4', num_classes=NUM_CLASSES, dropout_rate=DROPOUT).to(device)
    criterion = nn.CrossEntropyLoss(label_smoothing=LABEL_SMOOTH)
    optimizer = Adam(model.parameters(), lr=LR)
    scheduler = CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS*len(train_loader), eta_min=1e-6)

    best_val_acc = 0.0
    best_epoch = 0
    epochs_no_improve = 0

    run_train_losses = []
    run_val_accs = []

    for epoch in range(1, NUM_EPOCHS + 1):
        model.train()
        running_loss = 0.0
        for imgs, labels in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            scheduler.step()
            running_loss += loss.item()

        model.eval()
        preds, truths, probs = [], [], []
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                outputs = model(imgs)
                p = torch.softmax(outputs, dim=1)
                _, predicted = torch.max(p, 1)
                preds.extend(predicted.cpu().numpy())
                truths.extend(labels.numpy())
                probs.extend(p.cpu().numpy())

        val_acc = accuracy_score(truths, preds)
        avg_loss = running_loss / len(train_loader)
        print(f"Run {run} Epoch {epoch} - Loss {avg_loss:.4f} - Val Acc {val_acc:.6f}")

        run_train_losses.append(avg_loss)
        run_val_accs.append(val_acc)

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_epoch = epoch
            epochs_no_improve = 0
            torch.save(model.state_dict(), PREPROCESSED_PATH / 'models' / f'run{run}_best_epoch{epoch}.pth')
        else:
            epochs_no_improve += 1

        if epochs_no_improve >= PATIENCE_ES:
            print(f"Run {run} - EarlyStopping at epoch {epoch}")
            break

    print(f"Run {run} Best Val Acc: {best_val_acc:.6f} at epoch {best_epoch}")

    # -------------------------------------------------------
    # Métricas y curvas sobre VALIDACIÓN  <<<<<
    # -------------------------------------------------------
    model.load_state_dict(torch.load(PREPROCESSED_PATH / 'models' / f'run{run}_best_epoch{best_epoch}.pth'))
    model.eval()

    val_preds, val_truths, val_probs = [], [], []
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs = imgs.to(device)
            outputs = model(imgs)
            p = torch.softmax(outputs, dim=1)
            _, predicted = torch.max(p, 1)
            val_preds.extend(predicted.cpu().numpy())
            val_truths.extend(labels.numpy())
            val_probs.extend(p.cpu().numpy())

    compute_metrics(val_truths, val_preds, val_probs,
                    class_names=list(mapping),
                    save_prefix=str(PREPROCESSED_PATH / 'results' / f'run{run}_val_'))

    errors = np.where(np.array(val_truths) != np.array(val_preds))[0]
    print(f"Total errores en val: {len(errors)}")
    val_dataset_vis = torch.utils.data.Subset(dataset, val_indices)
    val_dataset_vis.dataset.transform = None
    for i, idx in enumerate(errors[:5]):
        img = val_dataset_vis[idx][0]
        img = (img.numpy().transpose(1, 2, 0) * 255).astype(np.uint8)
        plt.imsave(PREPROCESSED_PATH / 'results' / f'run{run}_val_error{i}_true{mapping[val_truths[idx]]}_pred{mapping[val_preds[idx]]}.png', img)

    # -------------------------------------------------------
    # Curvas de aprendizaje
    # -------------------------------------------------------
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(run_train_losses) + 1), run_train_losses, label='Train loss')
    plt.title(f'Run {run} – Loss por epoch')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(run_val_accs) + 1), run_val_accs, label='Val accuracy', color='orange')
    plt.title(f'Run {run} – Accuracy de validación por epoch')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True)
    plt.tight_layout()
    plt.savefig(PREPROCESSED_PATH / 'results' / f'run{run}_learning_curves.png')
    plt.close()

    report_lines = [
        f"Run {run} – Resumen de entrenamiento",
        "=" * 45,
        f"Hiper-parámetros fijados:",
        f"  - LR inicial:           {LR}",
        f"  - Batch size:           {BATCH_SIZE}",
        f"  - Label smoothing:      {LABEL_SMOOTH}",
        f"  - Dropout (Efficient):  {DROPOUT}",
        f"Técnicas de regularización usadas:",
        f"  • Dropout en EfficientNet (p={DROPOUT})",
        f"  • Label smoothing (ε={LABEL_SMOOTH})",
        f"  • Data augmentation (geometría + color + blur)",
        f"Validación cruzada: no utilizada.",
        f"  – Razón: split simple 85/15 para acelerar 50 runs.",
        f"  – Si se desea: reemplazar random_split por StratifiedKFold.",
        f"Resultados finales:",
        f"  – Mejor val accuracy:   {best_val_acc:.6f} (época {best_epoch})",
        f"  – Loss final (train):   {run_train_losses[-1]:.4f}",
    ]
    report_path = PREPROCESSED_PATH / 'results' / f'run{run}_report.txt'
    with open(report_path, 'w') as f:
        f.write("\n".join(report_lines))
    print(f"Run {run} – Reporte y curvas guardadas.")

    # -------------------------------------------------------
    # Inferencia sobre el TEST CIEGO  (solo para submission)
    # -------------------------------------------------------
    test_dataset = NumpyDataset(PREPROCESSED_PATH / 'X_test_B4_enhanced.npy', transform=None)
    test_dataset.transform = tta_transform
    test_loader = DataLoader(test_dataset, batch_size=max(1, BATCH_SIZE//5), shuffle=False, num_workers=4)

    y_pred = []
    for batch in test_loader:
        imgs = batch.to(device) if isinstance(batch, torch.Tensor) else batch[0].to(device)
        bs, ncrops, c, h, w = imgs.size()
        with torch.no_grad():
            outputs = model(imgs.view(-1, c, h, w))
            outputs = outputs.view(bs, ncrops, -1).mean(1)
            probs = outputs.softmax(dim=1)
            pred = probs.argmax(1).cpu().numpy()
            y_pred.extend(pred)

    submission_df = pd.DataFrame({'id': ids, 'label': [mapping[i] for i in y_pred]})
    submission_df.to_csv(PREPROCESSED_PATH / 'results' / f'submission_run{run}_acc{best_val_acc:.6f}_B4.csv', index=False)
    print(f"Run {run} submission: submission_run{run}_acc{best_val_acc:.6f}_B4.csv")
````

Aqui realizan múltiples entrenamientos (RUNS) de un modelo EfficientNet-B4 para clasificación de imágenes con PyTorch, utilizando un conjunto de entrenamiento y validación dividido previamente. Para cada ejecución, se entrena el modelo con técnicas de regularización como label smoothing, dropout, y data augmentation, ajustando los pesos mediante un optimizador Adam y un scheduler de tipo Cosine Annealing. Se aplican early stopping y se guarda el modelo con mejor precisión en validación. Luego, se calculan métricas y curvas de aprendizaje, se almacenan imágenes mal clasificadas, y se genera un reporte con los hiperparámetros y resultados. Finalmente, se realiza inferencia sobre un conjunto de test ciego usando Test Time Augmentation (TTA) y se guarda un archivo de submission con los resultados predichos.

#### Explicación hiperparametros

La configuración de los hiperparámetros en este entrenamiento está cuidadosamente balanceada para lograr una buena generalización sin sobreentrenar el modelo:

- Tasa de aprendizaje (`LR=8e-5`): Es lo suficientemente baja como para evitar saltos bruscos en la optimización, lo que ayuda a estabilizar el entrenamiento con modelos grandes como EfficientNet. Si se sube mucho, el entrenamiento se vuelve inestable y puede no converger. Si se baja demasiado, el modelo aprende muy lento y puede atascarse en mínimos locales.

- Batch size (`BATCH_SIZE=8`): Pequeño debido al alto consumo de memoria de EfficientNet-B4. Valores más altos estabilizan el gradiente, pero requieren más RAM. Más bajos pueden introducir mucho ruido y hacer que la convergencia sea errática.

- Label smoothing (`LABEL_SMOOTH=0.05`): Reduce la sobreconfianza del modelo suavizando las etiquetas verdaderas, lo que mejora la generalización. Si es muy alto, el modelo deja de aprender señales claras; si es cero, se arriesga a sobreajuste.

- Dropout (`DROPOUT=0.6`): Pensado para contrarrestar el sobreajuste por el alto poder de representación de EfficientNet. Bajarlo puede llevar a overfitting, subirlo demasiado puede impedir que el modelo aprenda bien.

- Early stopping (`PATIENCE_ES=7`): Da tiempo suficiente para recuperarse de pequeñas fluctuaciones sin alargar entrenamientos innecesarios.

#### Inciso: Fine tunning

Antes de llegar a esta versión final con EfficientNet, intentamos aplicar finetuning al modelo. En esa configuración, entrenamos primero durante 30 épocas congelando la base convolucional, y luego liberamos las capas superiores para realizar finetuning durante 20 épocas adicionales. Sin embargo, al evaluar esta versión en Kaggle, el puntaje obtenido fue considerablemente peor.

Una posible explicación para este descenso en rendimiento es que, al descongelar las capas preentrenadas de EfficientNet, el modelo comenzó a sobreajustarse al conjunto de entrenamiento. Esto puede ocurrir cuando los pesos preentrenados, que están ajustados a un dominio general (como ImageNet), se alteran demasiado durante el finetuning sin un control adecuado, como una tasa de aprendizaje más baja o regularización adicional. Como resultado, el modelo pierde la capacidad de generalizar bien al conjunto de prueba, degradando su rendimiento en métricas externas como las de Kaggle.

### Importación de librerias

````python
import pandas as pd
import numpy as np
from PIL import Image
import os
import json
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
````

### Implementación del LLM: Establecer promt

Como vamos a utilizar el LLM de google, necesitamos establecer un promt al que le vamos a enviar por cada imagen, indicando que realice la clasificación entre las diferentes categorias.

````python
prompt = """
    This is a brain magnetic resonance imaging (MRI) scan.
    Classify the image into one of the following categories:

    - Glioma
    - Meningioma
    - Pituitary
    - No tumor

    Respond only with the name of the corresponding class. Do not provide any additional explanation.

    """
````

### Limpiar la respuesta del LLM

Como el LLM puede retornar un mensaje más largo del que necesitamos, vamos a crear una función que limpie la respuesta y retorne la etiqueta predicha

````python
def clean_llm_output(output):
    output_clean = output.strip().lower()

    if "glioma" in output_clean:
        return "glioma"
    elif "meningioma" in output_clean:
        return "meningioma"
    elif "pituitary" in output_clean:
        return "pituitary"
    elif ("no tumor" in output_clean) or ("no mass" in output_clean) or ("no lesion" in output_clean) or ("no abnormality" in output_clean):
        return "notumor"
    else:
        # Si no se reconoce ninguna clase, guarda lo que dijo el LLM para análisis posterior
        return output.strip()
````

### Carga de imágenes

Como el LLM de gemini solo permite clasificar un número limitado de imagenes, vamos a estar solicitando la clasificación de 14 imágenes por cada API KEY, guardando dichas predicciones en un json, para asi poder clasificar varias resonancias.

````python
TRAIN_IMAGES_PATH = "/content/drive/Shareddrives/ISI/train"
CSV_PATH = "/content/drive/Shareddrives/ISI/train_labels.csv"
SAVE_PATH = "/content/drive/Shareddrives/ISI/LLM"
os.makedirs(SAVE_PATH, exist_ok=True)
N_IMAGES = 14  # Número de imágenes a procesar con el LLM

# Cargar etiquetas
df = pd.read_csv(CSV_PATH)

# Cargar predicciones anteriores si existen
output_file = os.path.join(SAVE_PATH, "predicciones_llm.json")

if os.path.exists(output_file):
    with open(output_file, 'r') as f:
        results = json.load(f)
else:
    results = []

# Evitar duplicados
procesadas = set([r['image_id'] for r in results])

# Filtrar imágenes no procesadas
df_remaining = df[~df['id'].isin(procesadas)]
df_sample = df_remaining.head(N_IMAGES)
````

### Clasificación con el LLM y guardado incremental

Vamos a usar el modelo de gemini 1.5 flash para la clasificación

````python
MODEL = genai.GenerativeModel('gemini-1.5-flash')

for idx, row in df_sample.iterrows():
    img_id = row['id']
    true_label = row['label']

    img_path = os.path.join(TRAIN_IMAGES_PATH, img_id)
    image = Image.open(img_path).convert("RGB")

    try:
        response = MODEL.generate_content([prompt, image])
        llm_pred = clean_llm_output(response.text)

        # Guardar resultado
        result = {
            "image_id": img_id,
            "true_label": true_label,
            "llm_prediction": llm_pred
        }

        results.append(result)

        # Guardar en JSON incrementalmente
        with open(output_file, 'w') as f:
            json.dump(results, f, indent=4)

        print(f"[{idx+1}] Imagen {img_id}: Real={true_label} | Pred={llm_pred}")

    except Exception as e:
        print(f"Error al procesar {img_id}: {e}")
print(f"\nProceso finalizado. Resultados guardados en {output_file}")
````

Convertimos los resultados del diccionario "results" en un dataframe para una mejor visualización

````python
results_df = pd.DataFrame(results)
print(results_df)
````

### Obtenemos las métricas

Cargamos los datos del dataframe "resultados"

````python
results_df = pd.DataFrame(results)

# Asegurarse de tener columnas correctas
results_df['true_label'] = results_df['true_label'].str.lower().str.strip()
results_df['llm_prediction'] = results_df['llm_prediction'].str.lower().str.strip()

labels = ['glioma', 'meningioma', 'pituitary', 'notumor']
results_df
````

### Accuracy global

````python
correct = np.sum(results_df['true_label'] == results_df['llm_prediction'])
total = len(results_df)
accuracy = correct / total

print(f"\nAccuracy global: {accuracy:.2f}")
````

El modelo predijo correctamente el 62% de las 42 imágenes, es decir, de las 42 imágenes enviadas al LLM, aproximadamente 26 imágenes fueron clasificadas correctamente y 16 se clasificaron mal.

### Recall macro y otras métricas

````python
from sklearn.metrics import recall_score

recall_macro = recall_score(results_df['true_label'], results_df['llm_prediction'], labels=labels, average='macro')

final_score = (recall_macro + accuracy) / 2

print(f"Recall macro: {recall_macro:.2f}")
print(f"Métrica compuesta: {final_score:.2f}")
````

El recall mide cuán bien detecta el modelo cada clase sin importar cuántos ejemplos tenga cada una. El recall macro es el promedio del recall por clase. En el caso de estudio de las resonancias, este recall nos dice cuantos positivos reales detectó el modelo, y esto es mas importante en medicina, ya que, por ejemplo, es mejor decirle a un paciente que tiene un tumor cuando no lo tiene (y luego descartar con un examen), que decirle que no tiene nada cuando sí tiene un tumor (y dejarlo sin tratamiento).

La métrica compuesta pondera el desempeño global y desempeño balanceado por clase. Como el accuracy y el recall macro son iguales, significa que el LLM tiene un desempeño equilibrado entre clases, sin favorecer a una sobre las demás.

- Pituitario: El LLM es miu preciso, pero se le escapan varios casos reales. Al tener un recall de 0.54, significa que se le escaparon 6 casos reales, muchos los clasificó como meningioma.
- No tumor: Buen desempeño en general, detecta bien los casos sin tumor y no se confunde mucho. Es la clase que mejor maneja el LLM.

### Curva ROC y AUC por clase

La curva ROC y el cálculo del AUC son métricas diseñadas originalmente para problemas de clasificación binaria, ya que comparan la tasa de verdaderos positivos contra la tasa de falsos positivos. Para poder aplicarlas en un problema multiclase como este, es necesario transformar las etiquetas en formato binario utilizando la estrategia de "uno contra todos". Esto significa que, para cada clase, se considera esa clase como "positiva" (1) y todas las demás como "negativas" (0).

````python
# Binarizar etiquetas para ROC
y_true_bin = label_binarize(results_df['true_label'], classes=labels)
y_pred_bin = label_binarize(results_df['llm_prediction'], classes=labels)

plt.figure(figsize=(8,6))
for i, label in enumerate(labels):
    try:
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_bin[:, i])
        auc = roc_auc_score(y_true_bin[:, i], y_pred_bin[:, i])
        plt.plot(fpr, tpr, label=f"{label} (AUC={auc:.2f})")
    except ValueError:
        print(f"Advertencia: Clase {label} no tiene ejemplos positivos o negativos. Se omite en curva ROC.")

plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curvas ROC por clase")
plt.legend()
plt.show()
````

La curva ROC (Receiver Operating Characteristic) muestra la relación entre:
- True Positive Rate (TPR)
- False positive Rate (FPR)

Las curvas con crecimiento rápido hacia la esquina superior izquiera como la de no tumor, significa que el modelo (en ests caso el LLM) detecta muchos positivos reales sin comenter muchos falsos positivos. Las curvas cerca a la diagonal punteada indican que el modelo no logra distingir esa clase adecuadamente, como lo es con el glioma.

El AUC es el Área bajo la curva ROC.
- AUC = 1.0 es un modelo perfecto
- AUC = 0.5 es un modelo sin capacidad de discriminación, como dejarlo al azar
- AUC < 0.5 es un modelo peor que al azar (generalmente indica un problema en el modelo o etiquetado)

Para cada clase tenemos:
- Glioma: AUC de 0.54 indica que el LLM casi no logra distinguir los gliomas de otras clases.
- Meningioma: AUC de 0.80 indica que el LLM es bueno detectandolos.
- Pituitario: AUC de 0.77 indica que el LLM los discrimina razonablemente, pero tiene algunos falsos positivos.
- No tumor: AUC de 0.88. Es la clase con mejor desempeño.
